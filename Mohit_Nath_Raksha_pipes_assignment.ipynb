```python
!pip install faker

from faker import Faker
import random

fake = Faker()

# 1. Products table with product_price column
product_names = [
    ("PVC Pipe", 10.99), ("CPVC Pipe", 12.99), ("uPVC Pipe", 15.99), ("HDPE Pipe", 18.99),
    ("PVC Fittings", 5.99), ("CPVC Fittings", 7.99), ("uPVC Fittings", 9.99), ("HDPE Fittings", 11.99),
    ("PVC Valve", 8.99), ("CPVC Valve", 10.99), ("uPVC Valve", 12.99), ("HDPE Valve", 14.99),
    ("PVC Elbow", 4.99), ("CPVC Elbow", 6.99), ("uPVC Elbow", 8.99), ("HDPE Elbow", 10.99)
]

products = [(idx + 1, product_name, price) for idx, (product_name, price) in enumerate(product_names)]

# 2. Customers table
countries = ["Singapore", "Middle East", "Europe", "India", "Russia", "Australia"]

# Define states before the loop where it's used
states_by_country = {
    "Singapore": ["Central Region", "East Region", "North Region", "North-East Region", "West Region"],
    "Middle East": ["Dubai", "Abu Dhabi", "Doha", "Riyadh", "Muscat"],
    "Europe": ["London", "Paris", "Berlin", "Rome", "Madrid"],
    "India": ["Delhi", "Mumbai", "Bangalore", "Chennai", "Kolkata"],
    "Russia": ["Moscow", "Saint Petersburg", "Novosibirsk", "Yekaterinburg", "Kazan"],
    "Australia": ["Sydney", "Melbourne", "Brisbane", "Perth", "Adelaide"]
}

states = [(idx + 1, state, country_id + 1) for country_id, country in enumerate(countries) for idx, state in enumerate(states_by_country.get(country))]

customers = []
for idx in range(700):
    country = random.choice(countries)
    state = random.choice(states_by_country[country])  # Randomly select a state based on the chosen country
    customers.append((idx + 1, fake.name(), fake.street_address(), fake.phone_number(), state, country))

# 3. Sales table
sales = [(idx + 1, random.choice(products)[0], random.choice(customers)[0], random.randint(1, 10), fake.date_between(start_date='-1y', end_date='today')) for idx in range(700)]

# 4. Categories table
categories = [(idx + 1, category) for idx, category in enumerate(["agriculture", "plumbing", "industrial", "bath", "water_tank"])]

# 5. Orders table
orders = [(idx + 1, random.choice(customers)[0], random.choice(products)[0], fake.date_between(start_date='-1y', end_date='today'), fake.date_between(start_date='-1y', end_date='today')) for idx in range(700)]

# 6. Payments table
payments = [(idx + 1, random.choice(orders)[0], random.uniform(10, 1000), fake.date_between(start_date='-1y', end_date='today')) for idx in range(300)]

# 7. Suppliers table
suppliers = [(idx + 1, fake.company(), random.choice(countries)) for idx in range(100)]

# 8. Distributors table
distributors = [(idx + 1, fake.company(), fake.street_address(), fake.phone_number(), random.choice(countries)) for idx in range(50)]

# Print the generated data for each table
print("Products:")
for product in products:
    print(product)

print("\nCustomers:")
for customer in customers:
    print(customer)

print("\nSales:")
for sale in sales:
    print(sale)

print("\nCategories:")
for category in categories:
    print(category)

print("\nOrders:")
for order in orders:
    print(order)

print("\nPayments:")
for payment in payments:
    print(payment)

print("\nSuppliers:")
for supplier in suppliers:
    print(supplier)

print("\nDistributors:")
for distributor in distributors:
    print(distributor)
```

```python
import csv
import os

# Specify the directory path
output_directory = "/content/drive/My Drive/raksha_pipes_csv"

# Create the directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Save Products data to CSV


products_csv_path = os.path.join(output_directory, "products.csv")
with open(products_csv_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["product_id", "product_name", "price"])
    writer.writerows([(idx + 1, name, price) for idx, (name, price) in enumerate(product_names)])

# Rest of the code to save other tables' data to CSV

# Save Customers data to CSV
customers_csv_path = os.path.join(output_directory, "customers.csv")
with open(customers_csv_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["customer_id", "customer_name", "address", "phone_number", "state", "country"])
    writer.writerows(customers)

# Save Sales data to CSV
sales_csv_path = os.path.join(output_directory, "sales.csv")
with open(sales_csv_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["sale_id", "product_id", "customer_id", "quantity_sold", "sale_date"])
    writer.writerows(sales)

# Save Categories data to CSV
categories_csv_path = os.path.join(output_directory, "categories.csv")
with open(categories_csv_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["category_id", "category_name"])
    writer.writerows(categories)

# Save Orders data to CSV
orders_csv_path = os.path.join(output_directory, "orders.csv")
with open(orders_csv_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["order_id", "customer_id", "product_id", "payment_id", "order_date", "delivery_date"])
    writer.writerows(orders)

# Save Payments data to CSV
payments_csv_path = os.path.join(output_directory, "payments.csv")
with open(payments_csv_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["payment_id", "order_id", "amount", "payment_date"])
    writer.writerows(payments)

# Save Suppliers data to CSV
suppliers_csv_path = os.path.join(output_directory, "suppliers.csv")
with open(suppliers_csv_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["supplier_id", "company_name", "country"])
    writer.writerows(suppliers)

# Save Distributors data to CSV
distributors_csv_path = os.path.join(output_directory, "distributors.csv")
with open(distributors_csv_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["distributor_id", "company_name", "address", "phone_number", "country"])
    writer.writerows(distributors)

print("CSV files saved successfully.")
```
